# Swiss Electricity Production Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## SFOE Daily Production by Source (2024–2025)

```python
# Load data
df_sfoe = pd.read_csv("data/sfoe_daily_2024_2025.csv", parse_dates=["Date"])

# Plot all energy sources over time
fig, ax = plt.subplots(figsize=(14, 6))

for col in df_sfoe.columns[1:]:
    ax.plot(df_sfoe["Date"], df_sfoe[col], label=col, alpha=0.8)

ax.set_xlabel("Date")
ax.set_ylabel("Production (GWh)")
ax.set_title("Swiss Daily Electricity Production by Source (2024\u20132025)")
ax.legend(loc="upper left")
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

## Generation per Type – Daily Average (2024–2025)

```python
# Load data
df_gen = pd.read_csv("data/generation_per_type_24_25.csv", index_col=0)
df_gen.index = pd.to_datetime(df_gen.index, utc=True)

# Resample to daily mean for a cleaner plot
daily = df_gen.resample("D").mean()

# Plot
fig, ax = plt.subplots(figsize=(14, 6))

for col in daily.columns:
    ax.plot(daily.index, daily[col], label=col, alpha=0.8)

ax.set_xlabel("Date")
ax.set_ylabel("Generation (MW)")
ax.set_title("Swiss Hourly Generation by Type \u2013 Daily Average (2024\u20132025)")
ax.legend(loc="upper left")
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```
